```python
import pyttsx3
```

```python
engine = pyttsx3.init()
```

```python
def speak(text):
    engine.say(text)
    engine.runAndWait()
```

```python
def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for a command...")
        audio = recognizer.listen(source)
        
        try:
            print("Recognizing speech...")
            command = recognizer.recognize_google(audio, language='ar-EG')  # Arabic (Egypt)
            print("Recognized command: ", command)
            return command.lower()
        except sr.UnknownValueError:
            print("Speech was not understood.")
            return None
        except sr.RequestError:
            print("Could not connect to Google API.")
            return None
```

```python
def execute_command(command):
    if command:
        if 'ابدأ' in command:
            speak("Patrol started.")
        elif 'توقف' in command:
            speak("Patrol stopped.")
        elif 'افحص المنطقة' in command:
            speak("Scanning the area.")
        else:
            speak("Sorry, I did not understand the command.")
    else:
        speak("No command recognized.")
```

```python
import speech_recognition as sr

def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Calibrating for background noise...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        print("Listening...")
        audio = recognizer.listen(source)

    try:
        command = recognizer.recognize_google(audio, language="en-US")  # English only
        return command
    except sr.UnknownValueError:
        return "Sorry, I didn't catch that."
    except sr.RequestError:
        return "Error: Could not connect to the speech recognition service."

if __name__ == "__main__":
    while True:
        command = recognize_speech()
        print(f"You said: {command}")
```
